```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset
df= pd.read_csv("D:\\data analysis\\Telco Customer Churn Dataset\\WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
print(df.describe())
```

```python
#distribution of the target variable (Churn)
print(df['Churn'].value_counts())
```

```python
# Visualize the Churn distribution
df['Churn'].value_counts().plot(kind='bar', color=['green', 'red'], title='Customer Churn Distribution')
plt.xlabel('Churn')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(x='Churn', y='tenure', data=df)
plt.title('Tenure vs Churn')
plt.show()
```

```python
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title('Contract Type vs Churn')
plt.show()
```
